```python
import csv
import codecs
import prettytable

p1 = prettytable.PrettyTable(["A","B","C","D","E"], encoding = "big5")
p1.align["A"] = "l"

with codecs.open("student.csv","r","big5") as f:
  ret = list(csv.reader(f))
  print(ret)
  for dd in ret:
    #print(dd)
    p1.add_row([dd[0],dd[1],dd[2],dd[3],dd[4]])
print(p1)
```

```python
import requests
import io
import prettytable
import csv

r1 = requests.get(
      "		https://data.tycg.gov.tw/opendata/datalist/datasetMeta/download",
      headers = {
            "user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36",
            "referer" : "https://www.google.com/",
      },
      params = {
            "id" : "e44cc432-b31a-408c-996b-219ecf76fe7b",
            "rid" : "848824bd-29e2-4376-ab7e-82144a286997"
      },
      data = {}
)
r1.encoding = "utf-8"
f = io.StringIO(r1.text)
ret = list(csv.reader(f))
p1 = prettytable.PrettyTable(ret[0], encoding = "utf-8")
for dd in ret[1:]:
  p1.add_row(dd)
print(p1)
```

```python
import requests
import io
import prettytable
import json

r1 = requests.get(
      "		https://ws.hsinchu.gov.tw/001/Upload/1/opendata/8774/550/e2cd2cb1-3f26-4a38-97d1-77f3c901d34b.json",
      headers = {
            "user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36",
            "referer" : "https://www.google.com/",
      },
      params = {},
      data = {}
)

ret = json.loads(r1.text)
# print(ret[5]["辦理單位"])
# print(ret[3]["服務項目"])
p1 = prettytable.PrettyTable(list(ret[0].keys()), encoding = "utf-8")
for dd in ret:
  p1.add_row(list(dd.values()))
print(p1)
# for i in ret:
#   print(i["辦理單位"],i["許可日期"])
```
